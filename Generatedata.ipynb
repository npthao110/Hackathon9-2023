```python
# Define constants
MIN_PRODUCTION = 1000000  # 1 million
MAX_PRODUCTION = 3000000  # 3 million

# Create an empty DataFrame
data = pd.DataFrame(columns=[
    "Month of the Year", "Raw Material", "Packaging Material", "Energy Consumption",
    "Waste in Transportation", "Inflation", "Disposable Income", "Production", "Sold"
])

# Generate data for each month since 1960
for year in range(1960, 2023):
    for month in range(1, 13):
        production = random.randint(MIN_PRODUCTION, MAX_PRODUCTION)
        raw_material = round(production * 0.142)
        packaging_material = round(production * 0.028)
        energy_consumption = round(production * 0.3)
        waste_transportation = round(production * 0.025)
        inflation = round(random.uniform(0.01, 0.05), 2)
        disposable_income = round(random.uniform(2000, 5000))
        
        # Define rules for Sold column
        if month not in [5, 6, 7, 8, 12]:
            sold = random.randint(int(production * 0.4), int(production * 0.9))
        else:
            sold = random.randint(int(production * 0.7), int(production * 1.0))
        
        # Modify Sold based on inflation and disposable income
        sold *= (1 - inflation * 19)
        sold *= (1 + disposable_income / 10000)

        # Ensure Sold is not greater than Production
        sold = min(sold, production)

#         sold = production - inflation * 10000 + disposable_income * 100

        # Append data to the DataFrame
        data = data.append({
            "Month of the Year": month, "Raw Material": raw_material, "Packaging Material": packaging_material,
            "Energy Consumption": energy_consumption, "Waste in Transportation": waste_transportation,
            "Inflation": inflation, "Disposable Income": disposable_income,
            "Production": production, "Sold": round(sold)
        }, ignore_index=True)

# Save DataFrame to CSV
data.to_csv('new_file.csv', index=False)
```
